## Sales Analysis by Category

```python
sales_by_category = df.groupby("Category")["Sales"].sum().reset_index()
sales_by_category
```

```python
profit_by_category = df.groupby("Category")["Profit"].sum().reset_index()

profit_by_category
```

```python
category_analysis = df.groupby("Category")[["Sales", "Profit"]].sum().reset_index()

category_analysis["Profit Margin"] = (
    category_analysis["Profit"] / category_analysis["Sales"]
)

category_analysis
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    category_analysis["Category"],
    category_analysis["Profit"]
)

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```

```python
discount_by_category = df.groupby("Category")["Discount"].mean().reset_index()

discount_by_category
```

## Business Insights

```python
import numpy as np

x = np.arange(len(category_analysis["Category"]))

width = 0.35

plt.figure(figsize=(10,5))

plt.bar(
    x - width/2,
    category_analysis["Sales"],
    width,
    label="Sales"
)

plt.bar(
    x + width/2,
    category_analysis["Profit"],
    width,
    label="Profit"
)

plt.xticks(
    x,
    category_analysis["Category"]
)

plt.title("Sales vs Profit by Category")
plt.ylabel("Amount ($)")
plt.legend()
plt.savefig("../images/sales_vs_profit_category.png", bbox_inches="tight")
plt.show()
```

### Business Insight

Technology generates the highest sales and profit, making it the strongest-performing category.

Furniture achieves sales comparable to Technology and Office Supplies but generates significantly lower profit. This suggests that the category operates with lower profit margins or is heavily affected by discounts and operational costs.

Office Supplies maintains a good balance between sales and profitability, making it a stable contributor to the company's overall performance.
